```python
def linebreak():
    """
    Print a line break
    """
    print("\n\n")

def start_game():
    """
    Start the game
    """
    print("You wake up on a couch and find yourself in a strange house with no windows which you have never been to before.")
    print("You don't remember why you are here and what had happened before.")
    print("You feel some unknown danger is approaching and you must get out of the house, NOW!")

    # Import library
    from IPython.display import Image, display

    # Image URL
    url = "https://static.wixstatic.com/media/90cee8_9339dce0c79e4bf282bd35f2e5f63f7d~mv2.png/v1/fill/w_640,h_384,al_c,q_85,usm_0.66_1.00_0.01,enc_avif,quality_auto/90cee8_9339dce0c79e4bf282bd35f2e5f63f7d~mv2.png"

    # Shows the image
    display(Image(url=url))
    
    play_room(game_state["current_room"])

def play_room(room):
    """
    Play a room. First check if the room being played is the target room.
    If it is, the game will end with success. Otherwise, let player either 
    explore (list all items in this room) or examine an item found here.
    """
    game_state["current_room"] = room
    if(game_state["current_room"] == game_state["target_room"]):
        print("Congrats! You escaped the room!")

        # Import library
        from IPython.display import Image, display

        # Image URL
        url2 = "https://images-wixmp-ed30a86b8c4ca887773594c2.wixmp.com/f/6f65e125-3e68-4e58-bf42-4c7cfcea8ed4/dgnbuyg-60be73c2-66ca-468b-b734-5d046dcc8fff.png?token=eyJ0eXAiOiJKV1QiLCJhbGciOiJIUzI1NiJ9.eyJzdWIiOiJ1cm46YXBwOjdlMGQxODg5ODIyNjQzNzNhNWYwZDQxNWVhMGQyNmUwIiwiaXNzIjoidXJuOmFwcDo3ZTBkMTg4OTgyMjY0MzczYTVmMGQ0MTVlYTBkMjZlMCIsIm9iaiI6W1t7InBhdGgiOiJcL2ZcLzZmNjVlMTI1LTNlNjgtNGU1OC1iZjQyLTRjN2NmY2VhOGVkNFwvZGduYnV5Zy02MGJlNzNjMi02NmNhLTQ2OGItYjczNC01ZDA0NmRjYzhmZmYucG5nIn1dXSwiYXVkIjpbInVybjpzZXJ2aWNlOmZpbGUuZG93bmxvYWQiXX0.p3zlWLVadOOE90Q85qv-xlZ-zKo7_AUvKU-4gwPozss"

        # Shows the image
        display(Image(url=url2))





    
    else:
        print("You are now in " + room["name"])
        intended_action = input("What would you like to do? Type 'explore' or 'examine'?").strip()
        if intended_action == "explore":
            
            # Import library
            from IPython.display import Image, display

            # Image URL
            url1 = "https://www.shutterstock.com/image-photo/dora-explorer-thinking-her-hand-600nw-2527022851.jpg"

            # Shows the image
            display(Image(url=url1))

            explore_room(room)
            play_room(room)
        elif intended_action == "examine":

            # Import library
            from IPython.display import Image, display

            # Image URL
            url2 = "data:image/jpeg;base64,/9j/4AAQSkZJRgABAQAAAQABAAD/2wCEAAkGBxMTEhUSEhIQFRUVFRUVFRUWEBAVFhAVFhUWFxUXFRUYHSggGBolGxUVITEhJSkrLi4uFx8zODMsNygtLisBCgoKDg0OFxAQGC0dHR0rLS0rLS0rLS0tKy0tKy0tKy0rLS0tKy0tLS0tLS0tLS0tNzc3LS0tKysrKysrLSsrK//AABEIAKgBLAMBIgACEQEDEQH/xAAcAAABBQEBAQAAAAAAAAAAAAABAAIEBQYDBwj/xAA8EAABAwIEBAQDBQYGAwAAAAABAAIRAwQFEiExBkFRYRMicYEHkaEyQlKx8BQjYsHR4RUzQ3KC8WOS0v/EABgBAAMBAQAAAAAAAAAAAAAAAAABAgME/8QAIBEBAQEBAAIDAQEBAQAAAAAAAAECEQMhEjFBURNhIv/aAAwDAQACEQMRAD8A9iTSnoEIQYiCkQi0IBIpQlCASBRShIAilCUIAhEJLhd3TaYzOJ7ACS70H89kB3Qc4ASdgvNse+KlOm5zKAbUc2cxaQadKPx1vsk9mB3qvLuJviBfXwNI1HMo6y1nlzj+I9OyJOqma9V4r+LVvbuNK2b+0VBoSHRTYehfzPosPcfFTFHeYG3ptJ8oFGSewzGSvPqLg2Ggep6KVVxAnXTkB/CwdPUqvi2mMyN9R+KeI0oNc0HzByCm0OI9jothgfxes6pDbhtW3J+89ssP/Jsx7rwqvW11gndzon2Cj1K066p/ErnNfUDuMrd3+SfG7s0YB3edPYSpFviz3DODb5en74H0zHn7QvmjCcSc2GZiBuNTA9lf0sTdTcKgDHObqMzQQes/2WVtlP8Ayl+n0Xh98yswVKZBB6EGCNCD3BUoLy7gTHfH8SpTOWsx0uokj95TdEifvCZh3Lb19KsrttRuZp7EHRzD0cORTYWcqSkkUkiBApxTSnACbKcU1MFKcE2E5AGUCUUxyARK5uKJTCqgApAIopgE0p6agJKBSlFQDU4BKEQgEhCcUkA0oQnwgAgGwlCeQhCAjYjesoU3Vqrg1jAS4n9brwLjLjKtfPc0nwrfZtKmSX1Ry8R3Q75VsPjxjZZSpWjDGd3i1I3LW6NHu78l474vlzRtsPxf2Rxv4sz7qXUs8zNSGU27Mac0u5Zo3d2Gyr3ugREAcpO/eEDcO5mT1nRg6NA/kjcuIAJMdBG/yEIkrb0ju76DptKYXcyf1yXN1T9bpNE6nbmtWNvXfwyWl2vyTqTZaCdY0jVPbcEDJ93kenRcrckOkdde6SofXoZHb8gfYqTSvSRB/XqFPGG+Jrt90HrGw9VWYhhj6WpGnUbKPlm+qvlntLtb91J4qU3PY9uzmug7zB6jQadlvuGuOL65qBjcstBJdTbSbUPYAl2cdg0+y8rFYhPFaeTZGsxqPQp/FOuafQdhiVN7g26ovqVf/I+tUMDdzaDxTgjn4bXQtfwzdCpbtLXh4GYBwJPlDnBoM6ggQCDqIXy1Z3tXNLapk75nOdtzM/nuvZvh5xM5lw63vQ+nVeGNp1DlNO58uZoc9vldUynR27gROoU1jrHHqkpqaHTtt15JxRxmBQCKQCASSSIQAKYU4ppQDUIRRCoGwgnOTSmAKanEJhQElPTU4KAUIwiEkrQSCchCQAIwkigGwhKiX2J06TmscfO/Rjebjpt8/oVCveIbem5zKlakwsAL8z2jLOwjclM+V4z8cHk4hl1J8KkfbzR9S5YGtS0DeYj2jotf8Q8cZeXrq9P/AC2tbTYSNXhsnNHKS4rFXdcnTQacuev6+Sqe3XOTPsG1gD5TEfeiSP8AaFxrPzGZce7jumjQfJSG6ACJ0+RVo+3CrRj+ic0eX1XR4679E2m4E/yTHPZUW8j2+qmWNtNTKRPSOZ0/r9E1+0gbfku1m4+NS1Osk/IqLfS/jxrcCoeR7HRLHb9QQHAhW1zh7XtggGQqW1q5Xv7sb8wSP5haWzqaD2HquTX9a/TzTiPBjQcC2crtR26hUkr1niLDxWtyI1Elp9CV5TXpZSQeS6fFv5T2w8ueXsdbRwDmu1OUg6EA78pWvZetq0HUg8NrMJrUH/5ZqBs1AwkS0PY4HLoNHQCNliWqZQrkCOR1g9RzB5K9RM9vpXgPHalekBcNy12QKggA1ARpUy8jILT3BWsWS4Q/f06Nwc2bLmBc0B7Q6Za4iM7ecnXbda0Hl0UMKQCJRhCEdIEiiECgGlAolBANhJOQKYNKaUSgnAUphT0EwkJIowswIRQCKQJJGEkGaijCSAoOKuGmXbQc1Rj2RlfTcA5pBkEToCNV4Vxvc1G3Bpur17gMluerTYDI+0JBI0I3X0Tf0nOGVri2dyImOcSvnbia2y3FVozZfFf9ok+Uu3J6nVFbeL7Z+6p52BzTqNSPTRUbjqpVWsWPcGnST12XJ7pM9fqtczitXoZOZ7aKRTqagDTMR7f3UVz5McvWUQYPpsnYJeJNYDl8+ZUeF1a8HsU2qQOcnty9SlFX+utAl3lnkVLw0TUZH3Wkk9NIH5qvpBwBcBE+XXvor3DbcMBy6kjVxED2HNTu8i8dq1tXTUPZon1kkK8tqsD9aqgwxmhP4jI7jYT+furu2HNcmmq0YZbHZeZcSWmSq4d9F6XQMrK8dWEs8Qfd39FXi1zSNzsYam3f5KxsajDDajXGC3K5pggSZBHPcQd9FBpO59PzT7S4cxwIMHUezhB+i7L9Oaen0RwnjFera0P2a3IeGClUq1n5adMt8ry1hGaoZboB6SFuLK3yNDS4vdu5x3cTuYG3ovGvhhjALRb3F8+k1hmn+8toqZiSQS6S0kknr6L2e1oUwPJBG8zJd3J5lZMtR3QKKUJJNQKckUwYUEUEwBTXJxTSnACEIowmDYSToQRA7NTkgEQsxCSRSQYpFJJAJJJKUA10c+i+dePy+ldkEf6Zeejs73kyOcGQF9FFeEfFovfdNY2m8up0GtJyEDV73ESRrAEoa+L7eY3Ls2u3ZR2p7neZMc/WQtp6XTXjVPpCTpquz6TqrppU3RAnSRMa69Fa2GCEDzEAkesJXUgzntVdZmgPrPskxhOvIfr2VzRwGs4AODGtBkuzSXDkAOSkUsGYCWkvHMgPMELO+SNZhUWnmdJacrZiAXZj6hXFOXbiG9OZ9egXdtoGjKNBy7LpQZGh6LLW5WuZyOts4x26dFYUXQFUtuWt3P8AVRLzGJ0aCB6qPjadrT/4m1g1/sulSs24pOYRJIInkFjLepm87m1HNG5ax5aPVwELXYLWkDyFvrzCes/FHevMa1Ise5h0LSQUxkfen6LXcc4RlrNe0H96I0/EIn6EKXg/AYuNLe7tKrxuySHNP8dNwzbzyI7rqzvs65tTh/AtW0pljqtZtFxM52FrvDby8ZrgQ4k7Bo05r6CwC98akHiC3ZrwIFSNMwB2C864L+HbgXi+pUS1paWtZSY1jt9HHKHEj1jUL1WlTDQGtAAAgACAANAAFFZavRISRQISQSa5OTSiA1AooKgBQARRATAJJyBCQBAlIpspwRIBRCaiFNByQQRSMUkkkAkkkkAyoYEwT2G5XlPxR4jpMp1KYo13V9BTIoVAzK+JDnEQQC2cvovWCot7Ra8Fr2OcO06+kFEOfb45rB2Y5hrz7HvGyYQvRviLws6m64rU6NSlb0302tBd4hqOeB4lR7sxcN4GbkF524ALaVo0vDl2fAc38Luk6HX+q7/tcSczlE4HqfvzTP8AqMI9xqP5q3xfD3ScrBoD199Bv6aLDXJrldGL/wCSscZZ9l4MeqVze0y6WyCOqoL9tNgaaVVziR5gaYBY7mC3p7qPTr1CdDH+5H+UvuKm/wAaG4uQS0fxA/KV0qUHO1YY6yNE2wqSBI/sexVozuVlfTTvVDd2YaC4y7KCT37Kkyvfu0QeUkR02WzuqEg9Fn61zSpnVwcejdVr49WxG+fqwwajVIDKjyWZg7INiRAAPYRstZ4zW+Zxa0dSQF53W4lftSaGdzqVU3N3UqGXvc71OnyT/wAtavtlfJmeo9avq1KrTpVmva5tGq55cDoMrHSCvJa1wS81GktOYuBBILSTOhGytMHvXeBVoTpo+OoH2h+SrLq3y6jb8lfjx8WetdjdcKfFy+tYZWIuqQ5VCRUaP4anP/lK9r4Q48s8QEUamWpGtGpDag6wJhw7iV8pAdP+16RwNwC4ltxclzIIcxgJa6dw4kaj0T3JGfxfRNWq1u5hQ6uLMHU/RZqriJDQC4mBEkyfdUt7isc1ldCY63rMWYev0XendMds4Ly6hjBndW7MW0Smj142/SWew3FSRofYq1p4gOYV9Z2cTUkynVDtinlMiTSigUACmpxTVUEd0kYSWYJJJJBiimynIBJJJFAKEkkkHxk+LuHLm7DqdO5pW9F4AqhtvnqVmj7pc50BvaF5Txf8JnUXMFoaj2BozuqQXPqEgNbSY0czvMAdV9BKBjWLUbWk6vcPaxjBq48+zRzceifacr51uuAb/Dyy8qNpmnTfTLy2oJYHEAyDE7xIWvxG2B8zf+55rEfEb4g1cSfkbNO2afJSnV5/HUjc9BsFa4BxMxtowXLix7PIAQc1Ro+yQNzpp7KPLm2db+LX9csRw+ZMD5BZ+vb6rYV7hlRoc0iHCRqqx1i3VxnudgPdTnV/XR6R8PoxAVoRsq51+xg8gB7z/Vcm8QAHzUzHUEE/JFzaXy4h8Z0neRwLsplpEmJ3GiysLd4lVZc27/DMkDNHNpGuoWFJW/i+uOby/fQSSSC1ZpGH18lRrjtsf9p0P5qytsMqVqho02lzgY7Bp2JPooGF2Dq9RtNm53P4RzK9l4fsGUGAADNABdAl0CNVnvfxXmekXhXgqhagVKkVKu8nZn+0LQ3WJNaNwqjFcYDdJCymJYvM6rHtq5hoMQx3UgFUz751RwYyXOP07rLXF84mASSdh1W44WsRRZndq92volqc+2nJI0GEYKxjQ6p5nfQeithd026ZW/RZ68xWBus3e44Qd/qlP+I+Nr0gXQ3bCmUb8HdeY4PxCc2UnT1Woo3fOVUrPWONnQu9dCrGjiX4lhad/Dhqpv8AiBjRV1HxbinXadiF0KwlHEzMTr6q8wbF8zhTcd9j0KfU3C+QhFBNMd0k1GVIIooJlWqGiXGP5oN0KQKqq2K6+UD3TqOJj7w9wg+VapLjRuGu2IK7IKEkvOOK/i/aWr30aVOpXqsJa4AZGNcNCC52p9gvLuIPi3iNxIZUbbsP3aQ83u86/KE5m1T3viriihY0y+q5maPLT8RjXPP/ACOgXznxjxU6+q+JcVX1Mv8Al0acso0R0Bd5nHTV0arL3FZz3F73Oe47uc4ucfUnVc1pM8NJ/bXDSmG0x/CPMfV51UYydSZPU6k+6QTlUga7hK5zDwt3SA3vOn5qbxK9jqn7O15DKRh0f6lQface06D0WPwu7dSqB7DBaZnpKngkkk7kkn1JkrPWffXR4r1OdhjY8rnKC+1dPpzVph9XkVKrUwNVn8rLxtZ1V29A/aZo9u/R45grP3dPK4j3Wjt68PlQ+J6QzNe0ctfdXnV6x8mfSilEDkNzp6lNV5wfa+JctJAIYC4z8gtbeTrCRseE8HFvTzP0e7VxP3RyCmYnjZjLSkfxcyomLX+pHIaQqmpdZWkkSTt27rm529dGc8ivxK8dzJn1VS67PVG8qyTK4UaWZ0LeTkTrXbxf8M2eZ3iv5bBaytfwFnbauGNAChXmJnqsrLqqnpY4lie+qoLm7J5qPVrklCmyVpMyJuvyJ2G1DmBWwtsSgbrIWlTSG7qytmkbnUqNT2f41tve5jKuKVXSFnsNpwJKuKSyTUqyqS5/ZWdpVgh3Qg/VVFvo0nm4qdSenE16TSqZgHDmAU5UnDV8HMNMnVuo7hXS1ntz2cp8ogoFMqPABJ5IAXV0GCeZ2/msxeYrJkn+y7YveFwnsYHQFZG6qmVlvTXGV+y7nmuwuFnadxAUS6xeNJUytPi1RvxyKd/jtVo8r/Y6rDOxnuuNTHOpT7R/mofiPgNSrc1LuiwRUhz2N3a6AHEdQYlefuEaGfQ8l65/iBeJCz2O4SytLtGv5OHPseq1x5L9Ua8f8YJJdbq3dTcWuEEfX0XFbMuCEU1P5Jg6g4BwnY6H0OiubXUa+h9RuqIqxwu4LXTodpB2KnU7PTTx65Vs6k8eZgJjkugvXPafKQeatcMxKgYzHIejtvYqfXq0IOVzC4iIGpPyXNdWfcdUsZC2aSV0vG+IHDtA9lPNmGDzvbTB3LjrHQN3UQ4hQYYY11Q9XeVvy3K0nv2jVjLQrLAL/wAGrJMBwyk9J2KFzampULgA0OMx0POAuxsGU2F7pJA9p5LS2c4wmK0FxVB1JH/13VfcOc90NEnks8ys4CATHqtNwRSJNWq6S1oyiT946/kpufjOqnk76VdayI3T7dgar+/pCSqWu2ES9Xxzr11Ce5PqlcHFXGer0XOhISUGNUukyEFJ1Iw3RXVlvJWeFWCrayugo1Gk/jUUKisraqsrTv4VxYVDEnmsrCXwfoOy60KirTW5LuyqpHF7h14ab2vHI/Mc16BTeHAOGxEheWUq3mAW0wLF2tpBrjqCR7bj81eax3lpHFU2JXWbQbD6p93fZtG7cz1VbUeE7UzKHeNLvqs5iYLJdlJA6CStFXrAFQqxBWdbZ9MXccQsA2f38pH5qF+1tq6scD25j1C1uJeE1jiWtmDyC8nxV0VC5hymdI0RmS+mv3F7eVC0Kiub0zuV1pY6C3LVGv4gNCuVWkx+rCCtpnn2np9tidVv2XGFZ08Ta8eeQeo29ws/lLDqNPyXdumu7Sq1mDqxxKxbUZuCeRkaFZEiDCu6tSJ10VI8ySe6eYz2CIKCKtmK62Z84BMAmJPJMpUi4wFJZatG5k/JHTi2fc0GaAPquHXyN/qfopNhi4d5YbSPRggO9HbrOOmUW678v1oouZWk3Y0F1hocSWu176z7qsdRLTBCi/tDhs5w90x1287un1CmZ1F3eb+LNlWOig4hd5/KNh9SotSqTuVzVzP6jXk7ODK2HCFYfs728/Fk/wDqIWOU3DMRNF0jUHRw6+ndGp2Iz6rU3tXV3qqe4qIVMVpukyRPKEaFu6o3PHlJMeymem3eohJKb4ani2TTRVSlxFZTTnLq8QuRKZfSO/dd6Dzs1Nq09dOas7KkGCeanV4JEnDLEzmeVpLZ30VHaXQJ1OqsWXACxvtS2B1XdpVSy5XencKeEtaX2wVYtq/qVU2lYGSpPiJS8KzrQ1sSIMBRnXZ5pJKenycNubnQH2K4+N3SSTKMxxPcHKV5xd1DKSS08UXfpFcmtcRqCR6JJLoYpVO/ds6Heu67U7xoB8pHbkUkkDtQ69wXnK1p9BJJUdzCDBBBG4Igj1CSSSaACKSSocSLV8SujqqKSRuXjLnUqztARSQHGUkEkypIoJIIkUEkHCW7wi2NO3Yx++pjoHawkkst/S8OVywBV1V6KSMtEN2qTKMoJKqSbRoKRVGiSSk5fSofWIKnUr8opJ8T1IZfFSad+kklw1laX3JWjbpJJZ2G/9k="

            # Shows the image
            display(Image(url=url2))


            
            examine_item(input("What would you like to examine?").strip())
        else:
            print("Not sure what you mean. Type 'explore' or 'examine'.")
            play_room(room)
        linebreak()

def explore_room(room):
    """
    Explore a room. List all items belonging to this room.
    """
    items = [i["name"] for i in object_relations[room["name"]]]
    print("You explore the room. This is " + room["name"] + ". You find " + ", ".join(items))

def get_next_room_of_door(door, current_room):
    """
    From object_relations, find the two rooms connected to the given door.
    Return the room that is not the current_room.
    """
    connected_rooms = object_relations[door["name"]]
    for room in connected_rooms:
        if(not current_room == room):
            return room

def examine_item(item_name):
    """
    Examine an item which can be a door or furniture.
    First make sure the intended item belongs to the current room.
    Then check if the item is a door. Tell player if key hasn't been 
    collected yet. Otherwise ask player if they want to go to the next
    room. If the item is not a door, then check if it contains keys.
    Collect the key if found and update the game state. At the end,
    play either the current or the next room depending on the game state
    to keep playing.
    """
    current_room = game_state["current_room"]
    next_room = ""
    output = None
    
    for item in object_relations[current_room["name"]]:
        if(item["name"] == item_name):
            output = "You examine " + item_name + ". "
            if(item["type"] == "door"):
                have_key = False
                for key in game_state["keys_collected"]:
                    if(key["target"] == item):
                        have_key = True
                if(have_key):
                    output += "You unlock it with a key you have."
                    next_room = get_next_room_of_door(item, current_room)
                else:
                    output += "It is locked but you don't have the key."
            else:
                if(item["name"] in object_relations and len(object_relations[item["name"]])>0):
                    item_found = object_relations[item["name"]].pop()
                    game_state["keys_collected"].append(item_found)
                    output += "You find " + item_found["name"] + "."
                else:
                    output += "There isn't anything interesting about it."
            print(output)
            break

    if(output is None):
        print("The item you requested is not found in the current room.")
    
    if(next_room and input("Do you want to go to the next room? Enter 'yes' or 'no'").strip() == 'yes'):
        play_room(next_room)
    else:
        play_room(current_room)
```
